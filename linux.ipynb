#rm명령어
rm (remove) 명령어는 유닉스 및 리눅스 계열 시스템에서 파일이나 디렉터리를 삭제하는 데 사용됩니다. 주요 옵션으로 -r (하위 디렉터리 포함 삭제)과 -f (강제 삭제)가 있으며, 한 번 삭제한 파일은 복구가 어려우므로 사용 시 주의가 필요합니다.
#주요 사용법 및 옵션:
파일 삭제: rm [파일명] (예: rm test.txt)
파일 여러 개 삭제: rm [파일1] [파일2]
디렉터리 및 하위 내용 삭제: rm -r [디렉터리명]
강제 삭제 (확인 메시지 없음): rm -f [파일명]
디렉터리 강제 삭제: rm -rf [디렉터리명]
대화형 삭제 (확인 후 삭제): rm -i [파일명]
#주의사항:
rm -rf 명령은 시스템 중요 파일을 예고 없이 삭제할 수 있으므로, 특히 루트 디렉터리(/)에 사용하지 않도록 극도로 주의해야 합니다.


#cp명령어
cp (copy) 명령어는 리눅스/유닉스 환경에서 파일이나 디렉토리를 복사하는 데 사용됩니다. 원본 파일은 그대로 유지하며, 이름이나 위치를 바꾸어 새로운 복사본을 생성할 수 있습니다. 기본 구조는 cp [옵션] <원본> <대상>입니다.
#주요 활용법 및 옵션:
파일 복사: cp file1.txt file1_backup.txt (file1.txt를 file1_backup.txt로 복사)
디렉토리 복사: cp -r dir1 dir2 (디렉토리 및 하위 내용까지 모두 복사)
복사 시 묻지 않고 덮어쓰기: \cp -f source target (alias 설정 무시하고 강제 덮어쓰기)
파일 속성 유지 복사: cp -p source target (원본 파일의 권한, 시간 등 정보 유지)
복사 과정 확인: cp -v source target (복사 진행 상황 화면 출력)

#mv명령어
mv (move) 명령어는 리눅스 및 유닉스 시스템에서 파일이나 디렉토리를 다른 위치로 이동하거나, 이름을 변경할 때 사용하는 핵심 명령어입니다. 기본적으로 mv [옵션] [원본] [대상] 형식을 사용하며, 원본 파일은 삭제되고 대상 위치에 새로운 파일로 저장됩니다.
핵심 용도 및 사용법
이 영상에서 mv 명령어를 사용하여 파일을 이동하고 이름을 바꾸는 방법을 배우세요:

#이동
파일/디렉토리 이동: mv file1.txt /home/user/documents/ (file1.txt를 documents 폴더로 이동).
파일/디렉토리 이름 변경: mv old_name.txt new_name.txt (파일 이름을 old에서 new로 변경).
여러 파일 이동: mv file1.txt file2.txt /folder/ (여러 파일을 한 번에 폴더로 이동).
자주 사용하는 옵션
-i (interactive): 이동할 위치에 동일한 이름의 파일이 있을 경우, 덮어쓸지 물어봅니다.
-n (no-clobber): 기존 파일을 덮어쓰지 않습니다.
-v (verbose): 이동 상황을 화면에 표시합니다.
#주의사항
대상 위치에 동일한 이름의 파일이 있으면 별도 옵션 없이 mv 명령어를 실행할 경우, 원본 파일이 덮어쓰기 되어 복구할 수 없게 되므로 주의해야 합니다.

#chmod명령어
chmod(change mode) 명령어는 유닉스 및 리눅스 계열 환경에서 파일이나 디렉토리의 접근 권한(Permission)을 변경하는 데 사용되는 핵심 명령어입니다. 소유자, 그룹, 기타 사용자에 대해 읽기(r), 쓰기(w), 실행(x) 권한을 기호(Symbolic)나 숫자(Octal) 모드로 설정하여 보안을 제어합니다.

# 권한 설정 방식 (모드)
숫자 모드 (Absolute Mode - 8진수): 3자리 숫자를 사용하여 권한을 설정합니다 (예: 755, 644).
첫 번째 자리(소유자), 두 번째 자리(그룹), 세 번째 자리(기타 사용자).
예: chmod 755 filename -> 소유자(rwx=7), 그룹(r-x=5), 기타(r-x=5).
기호 모드 (Symbolic Mode): 대상(u, g, o, a), 연산자(+, -, =), 권한(r, w, x)을 사용하여 설정합니다.
대상:
(user),
(group),
(others),
(all).
연산자:
 (추가),
 (제거),
 (설정).
# 주요 명령어 예시

chmod -R 755 directory: 디렉토리와 하위 파일 전체의 권한을 755로 변경.

#chown명령어
chown(change owner)은 리눅스/유닉스 시스템에서 파일이나 디렉터리의 소유자(User) 및 소유 그룹(Group)을 변경하는 명령어입니다. 주로 슈퍼 사용자(root) 권한이 필요하며, chown [옵션] [사용자]:[그룹] [파일/디렉터리] 형식을 사용합니다. 하위 디렉터리까지 일괄 변경하려면 -R 옵션을 사용합니다.
#핵심 요약 및 사용법
기본 구조: chown [사용자] [파일] (소유자만 변경)
소유자 및 그룹 변경: chown [사용자]:[그룹] [파일]
하위 파일/디렉터리 포함 전체 변경: chown -R [사용자]:[그룹] [디렉터리]
#주요 예시
파일 소유자만 변경: chown user1 file.txt (file.txt의 소유자를 user1로 변경)
파일 소유자 및 그룹 변경: chown user1:group1 file.txt (소유자 user1, 그룹 group1로 변경)
디렉터리 및 하위 파일 전체 변경: chown -R user1:group1 /path/to/dir
그룹만 변경: chown :group1 file.txt (소유자는 유지하고 그룹만 변경)

#ln명령어
리눅스 ln 명령어는 파일 시스템 내에서 파일이나 디렉토리의 링크(바로가기)를 생성하는 도구입니다. 원본 파일을 가리키는 심볼릭 링크(-s, 바로가기)와 원본과 동일한 아이노드를 공유하는 하드 링크 방식이 있으며, 원본 삭제 시 심볼릭 링크는 깨지지만 하드 링크는 데이터가 유지됩니다.

#주요 명령어 및 예시
심볼릭 링크 생성 (가장 많이 사용): ln -s [원본파일] [링크명]
예: ln -s /etc/nginx/nginx.conf ~/nginx.conf
하드 링크 생성: ln [원본파일] [링크명]
예: ln original.txt hardlink.txt
#옵션 및 특징
-s (symbolic): 심볼릭 링크(Soft Link)를 생성합니다.
-f (force): 동일한 이름의 링크 파일이 존재할 경우 삭제하고 새로 생성합니다.
대상(Target): 파일뿐만 아니라 디렉토리에 대한 심볼릭 링크도 생성 가능합니다.
하드 링크 특징: 파일 복사(cp)와 달리 원본과 하드 링크는 같은 디스크 공간을 공유하므로, 하나를 수정하면 다른 하나도 변경됩니다.



#find명령어
리눅스 find 명령어는 파일 시스템에서 조건(이름, 크기, 시간, 권한 등)에 맞는 파일이나 디렉터리를 검색하는 강력한 도구입니다. 기본 구조는 find [경로] [옵션] [표현식]이며, -name (이름), -type (타입: f-파일, d-디렉터리), -size (크기) 등이 가장 많이 사용됩니다.
#핵심 옵션 및 사용법:
기본 검색: find . -name "test.txt" (현재 디렉터리에서 test.txt 찾기)
파일/디렉터리 구분: find /home -type d -name "data" (/home에서 data라는 디렉터리 검색)
크기 기준: find / -size +100M (100MB 이상인 파일 검색)
시간 기준: find . -mtime -7 (최근 7일 이내에 수정된 파일)
검색 후 명령 실행: find . -name "*.log" -delete (.log 파일 찾아 삭제)
권한 검색: find . -perm 777 (모든 권한이 부여된 파일 찾기)

#주요 활용 팁:
find는 하위 디렉터리까지 재귀적으로 검색합니다.
경로 생략 시 현재 위치(.)에서 검색합니다.
-exec 옵션을 사용하면 찾은 파일들에 대해 ls -l, rm, chmod 등의 명령을 바로 적용할 수 있습니다.
