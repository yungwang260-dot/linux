#ㅣs 명령어 설명
ls (list segments) 명령어는 리눅스/유닉스 환경에서 디렉토리 내부의 파일과 하위 디렉토리 목록을 화면에 출력하는 핵심 명령어입니다. 기본적으로 현재 디렉토리의 내용을 보여주며, 다양한 옵션을 통해 상세 정보, 숨김 파일, 정렬 방식 등을 제어할 수 있습니다.
#주요 ls 명령어 옵션

ls -l: 상세 정보(권한, 소유자, 크기, 수정 시간)를 롱 포맷으로 표시.
ls -a: 숨김 파일(.)을 포함한 모든 파일 표시.
ls -al (또는 ls -la): 숨김 파일을 포함하여 모든 파일의 상세 정보를 표시 (가장
ls -t: 파일의 수정 시간 순서대로 정렬.
ls -r: 목록을 역순으로 표시.
ls -R: 하위 디렉토리의 내용까지 재귀적으로 표시.

#cd명령어
cd (Change Directory) 명령어는 리눅스, 유닉스, 윈도우 등 운영체제 터미널에서 현재 작업 디렉토리(폴더)를 이동하는 가장 기본적인 명령어입니다. 주로 cd [이동할 디렉토리 경로] 형태로 사용하며, 절대 경로와 상대 경로를 모두 지원하여 편리하게 파일 시스템을 탐색할 수 있습니다.
#핵심 cd 명령어 및 팁:
cd: 홈 디렉토리($HOME)로 이동
cd ~: 홈 디렉토리로 이동
cd ..: 상위 디렉토리(부모 폴더)로 이동
cd -: 바로 이전 작업 디렉토리로 이동
cd /: 시스템의 최상위 루트 디렉토리로 이동
cd /path/to/dir: 절대 경로로 직접 이동
cd dir1/dir2: 상대 경로(현재 위치 기준)로 이동
cd는 chdir로도 불리며, 디렉토리 이동 외에도 스크립트나 배치 파일 내에서 작업 경로를 변경하는 데 필수적으로 사용됩니다.

#pwd명령어
