```python
import re

def format_text(text):
    sentences = re.split(r'(?<=[.!?])\s+', text)
    return "\n\n".join(sentences)  # 줄 간격 추가 (더 자연스럽게)


def improve_resume(resume, job):

    # 🔹 자연스러운 말투 버전
    keyword_sentence = {
        "책임": "저는 맡은 일은 끝까지 해내야 한다고 생각합니다. 그런 태도가 신뢰를 만든다고 믿고 있습니다.",
        "노력": "어떤 일이든 쉽게 포기하지 않고 꾸준히 노력하는 자세를 중요하게 생각합니다.",
        "협력": "혼자보다 함께할 때 더 좋은 결과가 나온다고 믿으며, 팀원들과의 협력을 중요하게 생각합니다.",
        "도전": "새로운 환경이나 어려운 상황에서도 피하지 않고 부딪혀보는 것을 중요하게 생각합니다.",
        "리더": "팀의 방향을 잡고 함께 이끌어가는 과정에서 책임감을 느끼며 성장해왔습니다.",
        "문제": "문제가 생기면 피하기보다 원인을 찾고 해결하려는 태도를 중요하게 생각합니다."
    }

    matched = []

    for key in keyword_sentence:
        if key in resume:
            matched.append(keyword_sentence[key])

    if not matched:
        matched.append("저는 다양한 경험을 통해 조금씩 성장해왔고, 그 과정에서 제 강점을 만들어왔다고 생각합니다.")

    # 🔹 자연스럽게 이어 붙이기
    final_sentence = " ".join(matched)

    # 🔹 줄바꿈 적용
    formatted_resume = format_text(final_sentence)

    return f"""
====================================
📄 개선된 자기소개서
====================================

{formatted_resume}

이러한 경험을 바탕으로 {job} 직무에서도 맡은 역할을 책임감 있게 수행하며 꾸준히 성장해나가고 싶습니다.

====================================
📌 예상 면접 질문
====================================

1. 위 경험 중 가장 기억에 남는 순간은 언제인가요?
2. 해당 경험을 {job} 직무에 어떻게 활용할 수 있다고 생각하나요?
3. 본인의 강점은 무엇이라고 생각하나요?
"""


# 실행
resume = input("자소서 또는 키워드를 입력하세요: ")
job = input("지원 직무를 입력하세요: ")

result = improve_resume(resume, job)

print("\n=== 결과 ===")
print(result)
```
